```python
config = DQNConfig()
config = config.environment(env="ALE/Pong-ram-v5")
config = config.framework("torch")
config = config.training(dueling=False)
config = config.training(double_q=False)
config = config.training(noisy=False)
config = config.training(gamma=0.99)
config = config.training(lr=1e-4)

replay_config = {
        "type": "MultiAgentPrioritizedReplayBuffer",
        "capacity": 50000,
        "prioritized_replay_alpha": 0.5,
        "prioritized_replay_beta": 0.5,
        "prioritized_replay_eps": 3e-6,
    }

config = config.training(replay_buffer_config=replay_config)
config = config.resources(num_gpus=1)
config = config.env_runners(num_env_runners=1)

stop_config = {"episode_reward_mean": 2.0, "time_total_s": 3600,}

tuner = tune.Tuner(
    "DQN",
    param_space=config,
    run_config=train.RunConfig(stop=stop_config),
)

results = tuner.fit()
#tune.run(
#    "DQN",
#    config=config,
#    stop=stop_config
#)
```
